```python
import pandas as pd
from sklearn import decomposition
import matplotlib.pyplot as plt
```

```python
def PCA(data):

    data.set_index(keys="Unnamed: 0", drop= True, inplace=True)

    sample_ids = data.columns

    X = [list(data[column]) for column in data.columns]
    set = ["GTEX" if "GTEX" in sample_id else "TCGA" for sample_id in sample_ids]

    pca = decomposition.PCA(n_components=2)
    pca.fit(X)
    X = pca.transform(X)

    colors = {'GTEX':'orange', 'TCGA':'green'}
    df = pd.DataFrame(dict(X=X[:,0], y=X[:,1], set=set))

    fig, ax = plt.subplots()

    ax.set_xlabel("PC1")
    ax.set_ylabel("PC2")
    ax.scatter(df["X"],df["y"],c=df["set"].map(colors), label='TCGA')

    ax.legend()

    fig.show()
```

```python

pre_data = pd.read_csv("D:/data/RNASeq/dataset/pre_removal.csv")
post_data = pd.read_csv("D:/data/RNASeq/dataset/post_removal.csv")
post_data.fillna(0)
```

```python
PCA(pre_data)
PCA(post_data)
```
